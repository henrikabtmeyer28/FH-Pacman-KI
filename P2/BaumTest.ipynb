```python
from Knoten import Knoten
```

```python
pacman_pos_x = 1
pacman_pos_y = 1
```

```python
level = [
    ["#", "#", "#", "#"],
    ["#", " ", "*", "#"],
    ["#", "*", "#", "#"],
    ["#", "#", "#", "#"]
]
```

```python
if __name__ == "__main__":
    start = Knoten(pacmanPos=(pacman_pos_x, pacman_pos_y), map=level)
    queue= [start]

    for i in range(10):

        aktuellerKnoten = queue.pop(0)
        kinderAktuellerKnoten = aktuellerKnoten.expand()
        queue.extend(kinderAktuellerKnoten)
        print(f"Knoten {i+1}: Position ({aktuellerKnoten.pacmanRow}, {aktuellerKnoten.pacmanColumn}), {len(kinderAktuellerKnoten)} Kinder")
        # Map ausgeben
        for reihe in aktuellerKnoten.map:
            print(reihe)
        print()
```
