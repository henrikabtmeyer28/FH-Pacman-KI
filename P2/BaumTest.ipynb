```python
from Knoten import Knoten
```

```python
level = [
    ["#", "#", "#", "#"],
    ["#", " ", "*", "#"],
    ["#", "*", "#", "#"],
    ["#", "#", "#", "#"]
]
```

```python
pacman_pos_x = 1
pacman_pos_y = 1
```

```python
if __name__ == "__main__":
    start = Knoten(level, (pacman_pos_x, pacman_pos_y))
    nodes = [start]
    expanded = []
    while len(expanded) < 10:
        current = nodes.pop(0)
        if current.isEndNode() is False:
            expanded.append(current)
            current.printLevel()
            for child in current.expand():
                nodes.append(child)
```
