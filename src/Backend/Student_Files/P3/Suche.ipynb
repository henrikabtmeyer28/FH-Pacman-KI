**Tipp:** Erstellt euch ein statisches Numpy Array mit einem Startknoten und einem Zielknoten, um die Funktionalität eurer Suchalgorithmen zu testen, so müsst ihr nicht immer den Server starten.

```python
from P3 import Knoten
```

```python
class Suche:

    def __init__(self, insert):
        self.openList = []
        self.closedList = []
        self.insert = insert

    def starte_Suchalgorithmus(self, node: Knoten) -> Knoten:
        """
        node: Startknoten
        @return: Zielknoten, oder Fehler (falls kein Zielknoten gefunden)
        """

        self.openList = self.insert(node, self.openList)

        while True:
            if not self.openList:
                return LookupError
            current = self.openList.pop(0)
            if (self.goal_test(current)):
                return current
            if (current not in self.closedList):
                self.closedList.append(current)
                for child in current.expand():
                    self.openList = self.insert(child, self.openList)

    def goal_test(self, node: Knoten) -> bool:
        for i in range(len(node.level)):
            for j in range(len(node.level[i])):
                if (node.level[i][j] == 1):
                    return False
        return True

    def construct_action_path(self, node: Knoten) -> []:
        current = node
        path = []
        while (current.parent is not None):
            difference_x = current.pacman_pos_x - current.parent.pacman_pos_x
            difference_y = current.pacman_pos_y - current.parent.pacman_pos_y
            action = None
            if difference_x == 0 and difference_y == 1:
                action = 3  # down
            if difference_x == 0 and difference_y == -1:
                action = 2  # up
            if difference_x == 1 and difference_y == 0:
                action = 1  # right
            if difference_x == -1 and difference_y == 0:
                action = 0  # left
            path.append(action)
            current = current.parent

        return path.reverse()
```
