```python
from Knoten import Knoten
```

```python
level = [
    ["#", "#", "#", "#"],
    ["#", " ", "*", "#"],
    ["#", "*", "#", "#"],
    ["#", "#", "#", "#"]
]
```

```python
pacman_pos_x = 1
pacman_pos_y = 1
```

```python
if __name__ == "__main__":
    startKnoten = Knoten(pacman_pos_x, pacman_pos_y, level)
    nodes = [startKnoten]

    for i in range(10):
        current = nodes.pop(0)
        extended = current.expand()
        nodes.extend(extended)
        print(f"Knoten {i+1}: Position ({current.pacman_pos_x}, {current.pacman_pos_y}), {len(extended)} Kinder")
        for row in current.level:
            print(row)
        print()
```
